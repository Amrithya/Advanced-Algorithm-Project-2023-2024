```python
def generate_increasing_matrix(M, N, start=1, step=1):
    return np.fromfunction(lambda i, j: start + step * (i + j), (M, N), dtype=int)
```

```python
import numpy as np
import random

def generate_positive_matrix(M_range, N_range, max_value=100):
    # Randomly select M and N within the specified ranges
    M = random.randint(*M_range)
    N = random.randint(*N_range)

    matrix = np.random.randint(1, max_value + 1, size=(M, N))

    start = random.randint(1, 100)

    m1 = generate_increasing_matrix(M, N, start)

    # Generate the matrix with random positive numbers

    return matrix, m1, m1[::-1, ::-1]


generated_matrix = generate_positive_matrix((0, 20), (0, 20))
for g in generated_matrix:
  print(g.shape, g)
```

```python
def manipulate_matrix(matrix, percent_zeros=0, percent_negatives=0, shuffle=False):
    manipulated_matrix = matrix.copy()
    total_elements = manipulated_matrix.size

    num_zeros = int(total_elements * percent_zeros / 100)
    zero_indices = np.random.choice(range(total_elements), num_zeros, replace=False)
    np.put(manipulated_matrix, zero_indices, 0)

    non_zero_indices = np.where(manipulated_matrix != 0)[0]
    num_negatives = int(len(non_zero_indices) * percent_negatives / 100)
    negative_indices = np.random.choice(non_zero_indices, num_negatives, replace=False)
    np.put(manipulated_matrix, negative_indices, -np.abs(manipulated_matrix[negative_indices]))

    if shuffle:
        np.random.shuffle(manipulated_matrix)

    return manipulated_matrix

def generate_manipulated_matrices(matrix):
  result = []

  result.append(-np.abs(matrix)) #fully_negative_matrix

  percentiles = [10, 25, 50, 75, 90]

  for p in percentiles:
    result.append(manipulate_matrix(matrix, percent_zeros=p)) #zeros_matrices
    result.append(manipulate_matrix(matrix, percent_zeros=p, shuffle=True)) #shuffled_zeros_matrices
    result.append(manipulate_matrix(matrix, percent_negatives=p)) #negatives_matrices
    result.append(manipulate_matrix(matrix, percent_negatives=p, shuffle=True)) #shuffled_negatives_matrices

    for p2 in percentiles:
      result.append(manipulate_matrix(matrix, percent_zeros=p, percent_negatives=p2)) #combined_matrices
      result.append(manipulate_matrix(matrix, percent_zeros=10, percent_negatives=25, shuffle=True)) #shuffled_combined_matrices

  return result
```

```python
def mass_Gen_TestCaseMatrices(ls):
  c = []
  for el in ls:
    temp = generate_manipulated_matrices(el)
    for t in temp:
      c.append(t)

  return c

grab = mass_Gen_TestCaseMatrices(L)
print(grab[0], len(grab))
```

```python
def efficient_reduce_matrices_list(matrices, target_length):
    if len(matrices) <= target_length:
        return matrices

    # Calculate variance for each matrix
    variances = [(i, np.var(matrix)) for i, matrix in enumerate(matrices)]

    # Sort by variance and select the top matrices
    variances.sort(key=lambda x: x[1], reverse=True)
    selected_indices = [i for i, _ in variances[:target_length]]

    # Return the list of matrices with the highest variances
    return [matrices[i] for i in selected_indices]

reduced_matrices = efficient_reduce_matrices_list(grab, 100)
```

```python
def sort_matrices_by_size(matrices):
    return sorted(matrices, key=lambda x: x.shape[0] * x.shape[1])

sorted_matrices = sort_matrices_by_size(reduced_matrices)
print(sorted_matrices[0])
```

```python
def maxMatrixSum_non_constraint(matrix):
    rows = len(matrix)
    cols = len(matrix[0])

    # Calculate cumulative sums
    cumulative_sums = [[0] * (cols + 1) for _ in range(rows + 1)]
    for i in range(1, rows + 1):
        for j in range(1, cols + 1):
            cumulative_sums[i][j] = (
                matrix[i - 1][j - 1]
                + cumulative_sums[i - 1][j]
                + cumulative_sums[i][j - 1]
                - cumulative_sums[i - 1][j - 1]
            )

    max_sum = float('-inf')
    top, left, bottom, right = 0, 0, 0, 0

    # Calculate submatrix sums using cumulative sums
    for i1 in range(1, rows + 1):
        for j1 in range(1, cols + 1):
            for i2 in range(i1, rows + 1):
                for j2 in range(j1, cols + 1):
                    submatrix_sum = (
                        cumulative_sums[i2][j2]
                        - cumulative_sums[i1 - 1][j2]
                        - cumulative_sums[i2][j1 - 1]
                        + cumulative_sums[i1 - 1][j1 - 1]
                    )

                    if submatrix_sum > max_sum:
                        max_sum = submatrix_sum
                        top = i1 - 1
                        left = j1 - 1
                        bottom = i2 - 1
                        right = j2 - 1
                        top_left = (top, left)
                        bottom_right = (bottom, right)
    return max_sum,top_left,bottom_right


def maxMatrixSum_constraint(matrix, K, L):
    rows = len(matrix)
    cols = len(matrix[0])

    # Calculate cumulative sums
    cumulative_sums = [[0] * (cols + 1) for _ in range(rows + 1)]
    for i in range(1, rows + 1):
        for j in range(1, cols + 1):
            cumulative_sums[i][j] = (
                matrix[i - 1][j - 1]
                + cumulative_sums[i - 1][j]
                + cumulative_sums[i][j - 1]
                - cumulative_sums[i - 1][j - 1]
            )

    max_sum = float('-inf')
    top, left, bottom, right = 0, 0, 0, 0

    # Calculate submatrix sums using cumulative sums
    for i in range(K, rows + 1):
        for j in range(L, cols + 1):
            submatrix_sum = (
                cumulative_sums[i][j]
                - cumulative_sums[i - K][j]
                - cumulative_sums[i][j - L]
                + cumulative_sums[i - K][j - L]
            )

            if submatrix_sum > max_sum:
                max_sum = submatrix_sum
                top = i - K
                bottom = i - 1
                left = j - L
                right = j - 1
    return max_sum,(top,left),(bottom,right)
```

```python
def generate_noncon_Dict(Lmat):
  ls = []
  for m in Lmat:
    dict = {}
    max_sum, top_left, bottom_right = maxMatrixSum_non_constraint(m)
    dict['matrix'] = m.tolist()
    dict["expected_result"] = max_sum
    dict["subarray_indices"] = [top_left, bottom_right]

    ls.append(dict)
  return ls

Constrained_Sols = generate_noncon_Dict(sorted_matrices)
```

```python
import numpy as np

def generate_rectangular_matrix(M, N):
    # Randomly generate numbers in the range [-100, 100]
    matrix = np.random.randint(-100, 101, size=(M, N))
    return matrix

# Generate a 20x20 matrix
rectangular_matrix = generate_rectangular_matrix(10, 10)
print(rectangular_matrix)
```

```python
import numpy as np

def generate_submatrix_tuples(matrix, max_k, max_l):
    tuples = []
    M, N = matrix.shape

    for K in range(1, min(M, max_k) + 1):
        for L in range(1, min(N, max_l) + 1):
            tuples.append((matrix, (K, L)))

    return tuples

def sort_tuples_by_submatrix_size(tuples):
    """
    Sort the list of tuples by the size of the submatrix (K * L).
    """
    return sorted(tuples, key=lambda x: x[1][0] * x[1][1])


# Generate and sort the tuples
submatrix_tuples = generate_submatrix_tuples(rectangular_matrix, int(rectangular_matrix.shape[0]/2), int(rectangular_matrix.shape[1]/2))
sorted_tuples = sort_tuples_by_submatrix_size(submatrix_tuples)

# You can now work with sorted_tuples which are sorted by submatrix sizes (K * L)
sorted_tuples
```
